### SQL & Python, deux langages qui "cohabitent"

```python
# IMPORTATION DES BIBLIOTHEQUES
import sqlite3
import pandas as pd

# CONNEXION EN LOCAL AVEC LA BD DANS LE MEME REPERTOIRE QUE LE NOTEBOOK
conn = sqlite3.connect('kata_movies.sqlite')

# FONCTION QUI EXECUTE LE RESULTAT ET CREE UN DATAFRAME CONTENANT LA REPONSE
def fn_SQL(query):
    global conn
    return pd.read_sql(query, conn)
```

```python
# EXECUTER UNE REQUETE

query = ''' 

SELECT 
    *
FROM
    movies

'''

fn_SQL(query)
```

Le résultat de notre requête peut être mise dans une variable Python.       
Ici, nous utilisons un dataframe(df). Variable de la bibliothèques pandas

```python
df = fn_SQL(query)
```

### Python et la manipulation de données

#### Récupération des données et création de datasets

```python
import requests
import pandas as pd 

# URL DE L'API GitHub POUR LISTER LES FICHIERS PRESENTS
url = f"https://api.github.com/repos/Vaelastraszz/Crime_pandas/contents/"

r = requests.get(url)

files = r.json()

if r.status_code == 200:

    dataframes = {}

    for url in pd.json_normalize(files)['download_url']:
        # print(url)
        if url.endswith('.csv'):
            # print(url)

            name = url.split('/')[-1].replace('.csv', '')

            dataframes[name] = pd.read_csv(url)

            print(f"le dataset '{name}' et son lien : dataframes[{name}]")
```

```python
# Je suis en mesure de récupérer l'adresse de téléchargement du fichier excel, [2] correspond à un élément que j'ai choisi aléatoirement.

pd.json_normalize(files)['download_url'][2]
```

```python
# Je vais isoleer le nom fu fichier .csv, en utilisant des méthodes python

pd.json_normalize(files)['download_url'][2].split('/')[-1]
```
